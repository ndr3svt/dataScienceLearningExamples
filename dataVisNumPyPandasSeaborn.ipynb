```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Method 1: Using pandas hist()
plt.figure(figsize=(10, 6))
df['Price'].hist(bins=5)
plt.title('Distribution of Book Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Method 2: Using seaborn distplot/histplot
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Price', bins=20)
plt.title('Distribution of Book Prices')
plt.show()

# Method 3: To see price ranges with value counts
price_ranges = pd.cut(df['Price'], bins=10)
price_distribution = price_ranges.value_counts().sort_index()

# Create bar plot of price ranges
plt.figure(figsize=(12, 6))
price_distribution.plot(kind='bar')
plt.title('Distribution of Book Prices')
plt.xlabel('Price Range')
plt.ylabel('Number of Books')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Get basic statistics of price
print(df['Price'].describe())
```

```python
#count plot for TYPE
import seaborn as sns
import matplotlib.pyplot as plt

# Create countplot
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Type')

# Customize the plot
plt.title('Count of Books by Type')
plt.xlabel('Book Type')
plt.ylabel('Count')

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Adjust layout to prevent label cutoff
plt.tight_layout()

plt.show()
```

```python
# SIMPLE STRIPPLOT FOR RATING AND TYPE
import seaborn as sns
import matplotlib.pyplot as plt

# Create stripplot
plt.figure(figsize=(10, 6))
sns.stripplot(data=df, x='Type', y='Rating')

# Customize the plot
plt.title('Distribution of Ratings by Book Type')
plt.xlabel('Book Type')
plt.ylabel('Rating')

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

# Adjust layout to prevent label cutoff
plt.tight_layout()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create stripplot with unique colors
plt.figure(figsize=(10, 6))
sns.stripplot(data=df, x='Type', y='Rating',
              hue='Type',  # This makes each type have a unique color
              palette='husl',  # Using 'husl' color palette for distinct colors
              jitter=True,  # Add jitter to reduce overlapping
              alpha=0.7)  # Slight transparency

# Customize the plot
plt.title('Distribution of Ratings by Book Type')
plt.xlabel('Book Type')
plt.ylabel('Rating')

# Rotate x-axis labels
plt.xticks(rotation=45)

# Remove redundant legend since colors are already shown by Type
plt.legend([],[], frameon=False)

# Adjust layout to prevent label cutoff
plt.tight_layout()

plt.show()
```

```python
# simple box plot for rating


import seaborn as sns
import matplotlib.pyplot as plt

# Create boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='Rating')

# Customize the plot
plt.title('Distribution of Book Ratings')
plt.ylabel('Rating')

# Adjust layout
plt.tight_layout()

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create thin boxplot without fill
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='Rating',
            width=0.3,        # Make the box thinner
            color='white',    # White fill color
            linewidth=1.5,
            linecolor='orange')    # Line width of the box

# Customize the plot
plt.title('Distribution of Book Ratings')
plt.ylabel('Rating')

# Adjust layout
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='Rating',
            width=0.3,
            color='white',
            linewidth=1.5,
            fliersize=4,          # Size of outlier points
            whis=1.5,             # Length of whiskers (1.5 IQR is default)
            medianprops={"color": "orange"},  # Color of median line
            boxprops={"edgecolor": "black"}, # Color of box edges
            whiskerprops={"color": "black"}, # Color of whiskers
            capprops={"color": "black"},     # Color of caps
            flierprops={"markerfacecolor": "black"}) # Color of outliers

plt.title('Distribution of Book Ratings')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Create a dummy DataFrame with some sample data
data = {
    'ID': [101, 102, 103, 104, 105],
    'Category': ['Apple', 'Banana', 'Apple', 'Orange', 'Banana'],
    'Score': [88, 92, 85, 90, 87]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Convert the 'Category' column to categorical type
df['Category'] = df['Category'].astype('category')

print("\nDataFrame Info After Converting 'Category' to Categorical:")
print(df.info())

# Practice Operations:
# 1. List the categories
print("\nCategories:", df['Category'].cat.categories)

# 2. Count the frequency of each category
print("\nCategory Counts:")
print(df['Category'].value_counts())

# 3. Filter DataFrame rows based on a categorical value (e.g., Banana)
banana_df = df[df['Category'] == 'Banana']
print("\nFiltered DataFrame (Category = 'Banana'):")
print(banana_df)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create the dummy DataFrame
data = {
    'ID': [101, 102, 103, 104, 105],
    'Category': ['Apple', 'Banana', 'Apple', 'Orange', 'Banana'],
    'Score': [88, 92, 85, 90, 87]
}
df = pd.DataFrame(data)

# Convert the 'Category' column to a categorical type
df['Category'] = df['Category'].astype('category')

# Create a count plot to display the frequency distribution of the 'Category' variable
plt.figure(figsize=(8, 4))
sns.countplot(x='Category', data=df)
plt.title("Frequency Distribution of Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
import pandas as pd

# Define the first DataFrame
df1 = pd.DataFrame({
    'key': ['A', 'B', 'C'],
    'val1': [1, 2, 3]
})

# Define the second DataFrame
df2 = pd.DataFrame({
    'key': ['B', 'C', 'D'],
    'val2': [4, 5, 6]
})

# Perform a full outer join on the 'key' column
df_full_outer = pd.merge(df1, df2, on='key', how='outer')
print("Full Outer Join Result:")
print(df_full_outer)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style and random seed for reproducibility
sns.set(style="whitegrid")
np.random.seed(1)

# Create a sample 'auto' DataFrame with 100 observations.
auto = pd.DataFrame({
    'disp': np.random.uniform(100, 300, 100),  # displacement values between 100 and 300
    'acc': np.random.uniform(10, 30, 100)       # acceleration values between 10 and 30
})

# Example 1: Pairplot using columns 'disp' and 'acc'
sns.pairplot(auto[['disp', 'acc']])
plt.suptitle("Pairplot: disp and acc", y=1.02)
plt.show()



```

```python


# Example 2: Scatterplot using columns 'acc' and 'disp'
plt.figure()  # Start a new figure
sns.scatterplot(data=auto[['acc', 'disp']], x='acc', y='disp')
plt.title("Scatterplot: acc vs disp")
plt.xlabel("Acceleration")
plt.ylabel("Displacement")
plt.show()


```

```python
sns.pairplot(auto[['acc','disp']])
```

```python

# Example 3: Histplot for both 'disp' and 'acc'
# Melt the DataFrame so that we have one column for the variable name and one for its value.
auto_melt = auto[['disp', 'acc']].melt(var_name='variable', value_name='value')

plt.figure()
sns.histplot(data=auto_melt, x='value', hue='variable', multiple='dodge', bins=20)
plt.title("Histogram for disp and acc")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Example 4: Pairplot using columns 'acc' and 'disp'
sns.pairplot(auto[['acc', 'disp']])
plt.suptitle("Pairplot: acc and disp", y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Let's create a sample dataset similar to the auto dataset
data = {
    'wt': [2500 + i*10 for i in range(500)],  # weights from 2500-7500
    'cyl': [4]*200 + [6]*150 + [8]*100 + [3]*30 + [5]*20  # different cylinder counts
}
auto = pd.DataFrame(data)

# Create the stripplot
plt.figure(figsize=(12, 18))
sns.stripplot(x='wt', y='cyl', data=auto)

# Customize the plot
plt.title('Car Weight by Number of Cylinders')
plt.xlabel('Weight')
plt.ylabel('Number of Cylinders')

plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style and random seed for reproducibility
sns.set(style="whitegrid")
np.random.seed(42)

# Create a sample 'auto' DataFrame with 50 observations.
# 'wt' will represent the weight of the car (similar to displacement)
# 'cyl' will represent the number of cylinders.
auto = pd.DataFrame({
    'wt': np.random.uniform(1.5, 5.5, 50),  # weight values between 1.5 and 5.5
    'cyl': np.random.choice([4, 6, 8], 50)   # cylinder numbers randomly chosen from 4, 6, or 8
})

# Create a scatterplot for wt vs cyl
plt.figure(figsize=(8, 6))
sns.scatterplot(x=auto['wt'], y=auto['cyl'])
plt.title("Scatterplot of Weight vs. Cylinders")
plt.xlabel("Weight")
plt.ylabel("Cylinders")
plt.show()
```
